```python
import pandas as pd

# Завантаження даних
data = pd.read_csv('/Users/konstantin/Documents/КПІ/ДИПЛОМ/CSV/hiv_ukraine_data_all.csv')

# Видаляємо нерепрезинтативні регіони "Крим" та "м. Севастополь", також нам не потрібен рядок "Україна", адже там загальні дані
data = data[~data['Region'].isin(['м. Севастополь', 'Україна'])]

# Обираємо необхідні стовпці
selected_columns = ['Date', 'Region', 'HIV_Rate', 'AIDS_Rate']
clean_data = data[selected_columns]

clean_data.to_csv('CSV/cleaned_hiv_data.csv', index=False, encoding='utf-8')

# Розділяємо на періоди 2014-2021 та 2022-2024
clean_data['Year'] = pd.to_datetime(clean_data['Date']).dt.year  # Додаємо стовпець із роком

# Дані з 2014 по 2021 рік
data_2014_2021 = clean_data[clean_data['Year'] <= 2021].drop(columns=['Year'])
data_2014_2021.to_csv('hiv_2014_2021_cleaned.csv', index=False, encoding='utf-8')

# Дані з 2022 по 2024 рік (міграційні перекоси)
data_2022_2024 = clean_data[clean_data['Year'] >= 2022].drop(columns=['Year'])
data_2022_2024.to_csv('hiv_2022_2024_cleaned.csv', index=False, encoding='utf-8')
```
